```python
# 6/6/2018 - Exercise #1
import numpy as np

# Create evenly spaced values of type np.int32 within the interval (1, 20, 1)
a = np.arange(1, 20, 1, dtype=np.int32)
print(a)


def myfunc(start, end, step):
    while start <= end:
        yield start
        start += step
        
for x in myfunc(a[0], a[len(a)-1], 0.2):
    if x > 12 and x < 14:
        print(x)
```

```python
a = np.array([[2, 4, 6], [2, 23, 6], [64, 23, 55]], dtype=np.int16)
print(a)
print(a.size)
print(a.shape)
print(type(a))
print(a[2, 2]) # this is how we index a particular element in the array

b = a[0,:]
print(b)
print(b.shape)

b[2] = 88
print(b[2]) # => since b is a view of a (the base), this change will be reflected in a as well
print(a[0,2]) # => should be 88

print(a)
print(b)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# create an array "a" with 512 points in the range (-2*pi, 2*pi)
a = np.linspace(-np.pi*2, np.pi*2, 512, endpoint=True)

# 'sin' and 'cos' functions have the same number of points (512)
b_sin, c_cos = np.sin(a), np.cos(a)

# plot the results of the two functions above
plt.plot(a, b_sin)
plt.plot(a, c_cos)
plt.show()

# we can create a figure with a specific size and dpi (dots per inch)
plt.figure(figsize=(10,8), dpi=120)

# when two graphics are needed in the same graphic window, we can use 'subplot'
plt.subplot(3, 2, 6) 
# this will create an empty plotting space inside the window
# where 'y' is the y-axis, 'x' is the x-axis, and 
# 'n' is the number of the window to be created after setting x and y
plt.draw()
```

```python
import numpy as np
import matplotlib.pyplot as plt

d = np.linspace(-np.pi*3, np.pi*3, 128, endpoint=True)

d_sin = np.sin(d)
d_cos = np.cos(d)

# plot 'sin' using a green dash-dotted line of width 1.5px in area (2,1,1)
plt.subplot(2,1,1)
plt.plot(d, d_sin, color="red", linewidth=1.5, linestyle="-.")

# plot 'cos' using a blue dashed line of width 1.5px in area (3,2,6)
plt.subplot(3,2,6)
plt.plot(d, d_cos, color="blue", linewidth=2.5, linestyle="--")

# plt.show()  
# if we leave the above line without commenting it out, then the next cell will not have data since
# it seems as if plt.show() removes the data from the chart after showing.
```

```python
# this cell is sequential with the last side
# matplotlib - setting limits, ticks; showing and saving the plot

# we need to set the 'x' limits:
plt.xlim(-8.0, 8.0)
# then plot the 'x' ticks
plt.xticks(np.linspace(-8, 8, 6, endpoint=True))

# the above lines will apply to the plot in subplot (3,2,6), since it was the last active one
# to switch back, just do the following line:
# plt.subplot(2,1,1)

# now we set the 'y' limits 'y' ticks
plt.ylim(-1.2, 1.4)
plt.yticks(np.linspace(-1.2, 1.4, 4, endpoint=True))

# we can now save the figure using 64 dots per inch:
plt.savefig("Lecture_5.png", dpi=256)

# show the results on the screen
# plt.show()

# if we do plt.show() before plt.savefig(), an empty chart will save.  So, plt.show() most likely clears the memory
# for the specific chart
```

```python
import pylab as plb

%matplotlib notebook # this sets the mode of matplotlib

for i in range(10):
    k = 8
    x = plb.arange(k)
    y1 = plb.rand(k) * (1 - x / k)
    y2 = plb.rand(k) * (1 - x / k)
    plb.axes([0.075, 0.075, .88, .88])

    plb.bar(x, +y1, facecolor="#9922aa", edgecolor="green")
    plb.bar(x, -y2, facecolor="#ff3366", edgecolor="green")

    for a,b in zip(x, y1):
        plb.text(a+0.41, b+0.08, '%.3f' % b, ha='center', va='bottom')
    for a,b in zip(x, y2):
        plb.text(a+0.41, -b-0.08, '%.3f' % b, ha='center', va='top')

    plb.xlim(-.5, k), plb.ylim(-1.12, 1.12)
    plb.grid(True)
    plb.pause(1)
    plb.cla()
```

```python
import pylab as plb

plb.figure(num=1, figsize=(3,3))
gaus_dist = plb.normal(size=512) # create a random floating point vector
unif_dist = plb.uniform(-5, 5, size=512) # create uniform distribution vector

# plot the histogram with specific: bin number, color, transparency, label
plb.hist(unif_dist, bins=24, histtype='stepfilled', normed=True, color='cyan', label='Uniform') # normed is normalized, but check documentation
plb.hist(gaus_dist, bins=24, histtype='stepfilled', normed=True, color='orange', label='Guassian', alpha=0.5) # alpha sets the transparency


plb.legend(loc='upper left')
plb.title('Gaussian vs. Uniform Distribution / Histogram')
plb.xlabel("Value")
plb.ylabel("Frequency")
plb.grid(True)
plb.show()
```

```python
# Pie chart!
import pylab as plb

plb.figure('How do we get to work:', figsize=(3,3))
plb.axes([0.035, 0.035, 0.9, 0.9])
l = 'Car', 'Truck', 'Boat', 'Dingie', 'Train', 'Plane', 'Bus', 'Rocket', 'Tram', 'Other'
b = plb.round_(plb.random(10), decimals=2)
c = ['blue', 'red', 'green', 'gray', 'yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'cyan', 'orange']
e = (0, 0, 0, 0, 0, 0, 0, 0.05, 0, 0)

plb.cla()
plb.pie(b, explode=e, labels=l, colors=c, radius=0.75, autopct='%1.2f%%', shadow=True, startangle=0) # explode gives the pie chart the "important slice" effect
# startangle refers where the light source should be for the shadow to go in the opposite direction

# set the aspect ratio to 'equal' so that the pie is drawn in a circle
plb.axis('equal')
plb.xticks(()); plb.yticks(())
plb.show()

```

```python
# contour plot
import pylab as plb

def f(x,y):
    return (2 - x/3 + x**6 + 2.125*y) * plb.exp(-x**2 - y**2)

n = 128
x = plb.linspace(-2, 2, n)
y = plb.linspace(-1, 1, n)
A,B = plb.meshgrid(x, y)

plb.cla()
plb.axes([0.075, 0.075, 0.92, 0.92])

plb.contourf(A, B, f(A,B), 12, alpha=.50, cmap=plb.cm.gist_rainbow)
C = plb.contour(A, B, f(A,B), 8, colors='black', linewidths=0.75) # lindwidths is the thickness of each contour line

plb.clabel(C, inline=1, fontsize=14) # this labels the contour lines with actual values
plb.xticks(())
plb.yticks(())
plb.show()
```

```python
# polar plot
import pylab as plb

a = plb.axes([0.065, 0.065, 0.88, 0.88], polar=True)

q = 24
t = plb.arange(0.015, 3*plb.pi, 3*plb.pi/q)
rad = 12 * plb.rand(q)
w = plb.pi / 4 * plb.rand(q)
ba = plb.bar(t, rad, width=w)

for r,bar in zip(rad, ba):
    bar.set_facecolor(plb.cm.jet(r/12.))
    bar.set_alpha(0.75)
    
plb.show()
```

```python
# 3d plot
import pylab as plb
from mpl_toolkits.mplot3d import Axes3D

ax = Axes3D(plb.figure())
x = plb.arange(-6, 3, 0.35)
y = plb.arange(-6, 6, 0.35)
x, y = plb.meshgrid(x, y)
k = plb.sqrt(x**2 + y**2)
z = plb.sin(k)

ax.plot_surface(x, y, z, rstride=2, cstride=1, cmap=plb.cm.gist_stern)
ax.contourf(x, y, z, zdir='z', offset=-3, cmap=plb.cm.gist_stern)
ax.set_zlim(-4,4) # sets z axis limits

plb.show()
```
