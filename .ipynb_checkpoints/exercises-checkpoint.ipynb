```python
# 6/6/2018 - Exercise #1
import numpy as np

# Create evenly spaced values of type np.int32 within the interval (1, 20, 1)
a = np.arange(1, 20, 1, dtype=np.int32)
print(a)


def myfunc(start, end, step):
    while start <= end:
        yield start
        start += step
        
for x in myfunc(a[0], a[len(a)-1], 0.2):
    if x > 12 and x < 14:
        print(x)
```

```python
# 6/6/2018 - Exercise #2 - Solution
class _file_operations():
    def write_my_file(A):
        file = open('exericse2', 'w')
        file.write(str(A))
        file.close()
        
    def read_my_file():
        file = open('exercise2', 'r')
        B = file.read()
        file.close()
        return B

B = np.random.randint(5, 45, 6)
B.tofile('exercise2', sep=',', format='%s')
B = _file_operations.read_my_file()
B = np.array(B.split(','), dtype=np.int16)
C = np.matrix([[3], [2], [5]])
D = C*B
_file_operations.write_my_file(D)
print(_file_operations.read_my_file())
```
