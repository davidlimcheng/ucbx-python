```python
# 6/6/2018 - Exercise #1
import numpy as np

# Create evenly spaced values of type np.int32 within the interval (1, 20, 1)
a = np.arange(1, 20, 1, dtype=np.int32)
print(a)


def myfunc(start, end, step):
    while start <= end:
        yield start
        start += step
        
for x in myfunc(a[0], a[len(a)-1], 0.2):
    if x > 12 and x < 14:
        print(x)
```
